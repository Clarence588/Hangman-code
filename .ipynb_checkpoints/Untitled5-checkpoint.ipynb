```python
import random 

def run_hangman(): # Function defining the whole game
    #introduction
    print(''' Welcome to Hangman game
You should guess the name of a fruit, a movie, or a country to save the man.
Let's go !!! ''')

    #lists of the names of animals, fruits, countries
    animals = ["dog","cat","zebra","lion","tiger","snakes","parrot","turtle","fish","cow","giraffe","rat","cow","bear","wolf"]
    fruits = ["apple","banana","lemon","strawberry","mango","kiwi","pineapple","avocado","cherry","grape","water melon","papaya","pear","orange","plum"]
    countries = ["Lebanon","Egypt","Germany","Brazil","Canada","Australia", "France", "Japan", "Greece","Portugal","India","Switzerland","Netherland","South Africa","Singapore"]
    car_brands = ["Toyota","BMW","Aston Martin","Tesla","Nissan","Kia","Ferrari","Porsche","Honda","Ford","Mercedes Benz","Suzuki","BYD","Hyundai","Subaru"]
    win_count = 0
    number_of_play = 0
    
    def gameplay(): #Function of the game process
        #asking the user for the type of word he/she wants to guess
        type_word = input("which one do you want to guess (animals/fruits/countries/car brands): ").lower()
        global word
        #choosing a random word from the choosen list
        while True:
            if type_word == "fruits":
                word = random.choice(fruits)
                break
            elif type_word == "countries":
                word = random.choice(countries).lower()
                break
            elif type_word == "animals":
                word = random.choice(animals)
                break
            elif type_word == "car brands":
                word = random.choice(car_brands).lower()
                break
            else:
                print("check your spelling and try again!")
                type_word = input("which one do you want to guess (animals/fruits/countries/car brands): ").lower()
            print("")
        
        
        print("It is a word that has "+ str(len(word)) + " letters")
        
        #hangman figure
        hangman = [
        """
         ------
         |    |
         |
         |
         |
         |
         |
         |
         |
        ----------
        """,
        """
         ------
         |    |
         |    O
         |
         |
         |
         |
         |
         |
        ----------
        """,
        """
         ------
         |    |
         |    O
         |   -+-
         | 
         |   
         |   
         |   
         |   
        ----------
        """,
        """
         ------
         |    |
         |    O
         |   -+-
         |  /
         |   
         |   
         |   
         |   
        ----------
        """,
        """
         ------
         |    |
         |    O
         |   -+-/
         |  /
         |   
         |   
         |   
         |   
        ----------
        """,
        """
         ------
         |    |
         |    O
         |   -+-/
         |  / |
         |   
         |   
         |   
         |   
        ----------
        """,
        """
         ------
         |    |
         |    O
         |   -+-/
         |  / |
         |    |
         |   | 
         |   | 
         |   
        ----------
        """,
        """
         ------
         |    |
         |    O
         |   -+-/
         |  / |
         |    |
         |   | |
         |   | |
         |  
        ----------
        """]
        global wrong, sofar, max_wrong
        
        wrong = 0                         
        max_wrong = 7                      
        used = []                          
        so_far = ("-") * len(word)       
        
        print("")
        
        while wrong < max_wrong and so_far != word: 
            print(" ")
            print(hangman[wrong])                
            print("Word so far: " + so_far)       
            print("Letters used: " + str(used))   
          
            
            guess = input("Enter your guess: ").lower()
          
            
            while guess in used:
                print("--- Try again... You've already used this letter ---")
                print(" ")
                guess = input("Enter your guess: ").lower()
                print(" ")
          
            
            used.append(guess)
          
            # checking the answer (TRUE or FALSE)
            if guess in word:
                print("--- True ! " + guess + " is a letter of the word ---")
                print(" ")
            
               
                new = ""        
            
                for i in range(len(word)):  
                    
                    if guess == word[i]:
                        new += guess     
                    else:                   
                        new += so_far[i] 
            
                
                so_far = new 
            
            else:
                print("--- False ! Try again ! ---")
                wrong += 1   
    
    # call gameplay function
    gameplay()
    
    if wrong == max_wrong:
        print(" ")
        print("The word is " + word)
        print("You didn't save the man :( , YOU LOSE ! ")
        number_of_play += 1
                
    else:
        print(" ")
        print("Correct ! The word is " + word)
        print("You have saved the man :) , YOU WIN !")
        number_of_play +=1
        win_count += 1 #count the number of wins
    while True:
        play_again = input("Do you want to play again: type no for quitting and yes to play again:") #ask if player want to play again
        if play_again == "yes":
            gameplay()
        elif play_again == "no":
            print("okay")
            print("this is a report")
            print("game plays",number_of_play)
            print("wins", win_count)
            break
        else:
            print("you typed it wrong")
            play_again = input("Do you want to play again: type no for quitting and yes to play again:") #ask if player want to play again          
# Call the function of the game itself
run_hangman()

#The whole function of the game is to check if the players letter inputs by comparing it to the randomly chosen word.
```
